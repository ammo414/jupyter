### 1)

In the consumption-leisure model, consider the Cobb-Douglas utility function:

$$u(c,l) = Ac^{\alpha}l^{1-\alpha}$$

where $A >  0$ is a positive constant and $0 < \alpha < 1.$ Suppose the government taxes the consumer’s labor income at rate $\tau_n \geq 0$. In other words, when the consumer works a fraction $n$ of
her time and earns labor income $wn$ she must pay $\tau_n wn$ to the government. Her after tax labor income is, therefore, $(1 − \tau_n) wn$. This question asks you to study the effects of a tax on labor
income, using the following steps:

#### a)
##### Write down the consumer's single budget constraint. Graph the constraint when there is no tax and when there is a positive tax rate.

The consumer's expenditure budget constraint is

$$ (1 -\tau_n) wn = pc$$


The time constraint is $$n = 1 - l$$

For ease of calculation, let $ (1 - \tau_n) = X $



Combined, this results in a single budget constraint of 

$$ pc = Xwn$$


<br/>


$$pc = Xw(1 - l)$$


<br/>



$$pc = Xw - Xwl \tag{0}$$



Graphically:

$$pc = w(1 - \tau_n) - wl(1 - \tau_n)$$

<br/>

$$c = (1 - \tau_n)\frac{w}{p} - (1 - \tau_n) \frac{wl }{p} $$


We can see that if $\tau_n$ = 0, then this function collapses into 

<br/><br/>


$$c = \frac{w}{p} - \frac{w}{p} l$$


Which, when graphed out, is

```python
import matplotlib.pyplot as plt
import random as rand
import numpy as np
```

```python
rwage = rand.randint(0,100)/100 # treat w/p as a constant, chosen randomly

cval = []
lval = np.linspace(0,1, 11) # find 10 values for l between 0 and 1

for l in lval:    
    cval.append((rwage-rwage*l)) #solve equation for 10 l values
    
plt.plot(lval, cval)
plt.ylabel('c')
plt.xlabel('l')
plt.title('Budget constraint with w/p = {}, tau = 0'.format(rwage))
plt.show()
```

Importantly, the y-intercept of this constraint is $w/p$, and the x-intercept is $1$.

If $\tau_n \geq 0$, then the graph looks like

```python
tau = rand.randint(0,100)/100 # treat tau as a constant, chosen randomly

cvaltau = []

for l in lval:
    cvaltau.append((1-tau)*rwage - (1-tau)*rwage*l)

plt.plot(lval, cvaltau)
plt.ylabel('c')
plt.xlabel('l')
plt.title('Budget constraint with w/p = {}, tau = {}'.format(rwage, tau))
plt.show()
```

We can see that the y-intercept in the case of $\tau_n \geq 0$ is $(1 - \tau_n) \frac{w}{p}$, and the x-intercept is still $1$. Combining these two graphs, we see

```python
plt.plot(lval, cval, label = 'tau = 0')
plt.plot(lval, cvaltau, label = 'tau = {}'.format(round(tau, 2)))

plt.ylabel('c')
plt.xlabel('l')
plt.title('Budget Constraint with w/p = {}'.format(round(rwage, 2)))

plt.legend(loc='upper right')
plt.show()
```

#### b)

##### Write down the consumer's optimization problem and the Lagrangian fucntion for solving this optimization problem.

The consumer is trying to optimize her utility function, $u(c,l)$, with the budget constraint of $Xw - Xwl - pc = 0$.

<br/>

$$L() = A\, c^{\alpha}l^{1-\alpha} + \lambda[Xw - Xwl - pc ] \tag{1}$$

<br/>


#### c) 

##### Derive the first order conditions for this problem.

We optimize the following three equations:

$$\frac{\partial L()}{\partial c} = A \, l^{1-\alpha} \times \frac{\partial}{\partial c} c^{\alpha} - \lambda p$$

<br/>



$$ = A \, \alpha \, c^{\alpha -1} l ^{1 - \alpha}- \lambda p$$

<br/>


$$ = A \, \alpha \, c^{- (1 - \alpha)} \, l^{1 - \alpha} - \lambda p$$

<br/>


$$= A \, \alpha \frac{l^{1-\alpha}}{c ^ {1 - \alpha} } - \lambda p \tag{2}$$

$$\frac{\partial L()}{\partial l} = A c^{\alpha} \times \frac{\partial}{\partial l} \, l^{1 -\alpha} - \lambda w$$

<br/>


$$ = (1 - \alpha) l ^{ 1 - \alpha - 1} \, A \, c^{\alpha} - \lambda Xw$$


<br/>

$$ = (1 - \alpha) l ^{ -\alpha} \, A \, c^{\alpha} - \lambda Xw $$


<br/>

$$ = A (1 - \alpha) \, \frac {c^{\alpha}}{l^{\alpha}} - \lambda Xw \tag{3}$$

$$ \frac{\partial L()}{\partial \lambda} = Xw - Xwl - pc  \tag{4}$$

#### d)

##### Solve to find the consumer's optimal choice of $(c, l)$


Eliminating the $\lambda$ in (1) and (2), we get

$$A \alpha \, \frac{l^{1 - \alpha}}{c^{1 - \alpha}} \frac{1}{p} = A(1 - \alpha) \frac{c^{\alpha}}{l^{\alpha}}\frac{1}{Xw}$$

<br/>


$$A \alpha \frac{l^{1 - \alpha} \,X w}{c^{1 - \alpha} \, p}  = A(1 - \alpha) \frac{c^{\alpha}}{l^{\alpha}} $$

<br/>


$$ \frac{Xw}{p} = \frac{(1 - \alpha)}{\alpha} \frac{c^{\alpha} \, c^{1 - \alpha}}{l^{1 - \alpha} \, l^{\alpha}}$$

<br/>


$$ \frac{Xw}{p} = \frac{(1 - \alpha)}{\alpha} \frac{c}{l} \tag{5}$$

<br/>


Solving for $cp$, we get

<br/>


$$ cp = Xw \, l \, \frac{\alpha}{(1 - \alpha)} $$

<br/>


$$ cp = (1 - \tau_n) w \, l \, \frac{\alpha}{(1 - \alpha)} \tag{6}$$



Substituting this equality for $cp$ in (4), we get

$$ Xw - Xwl - Xwl \frac{\alpha}{(1 - \alpha)} = 0$$

<br/>

$$ 1 - l - l \frac{\alpha}{(1 - \alpha)} = 0$$

<br/>

$$1 = l\left(1 + \frac{\alpha}{\left(1 - \alpha\right)}\right)$$





$$l = \frac{1}{\left(1 + \frac{\alpha}{(1 - \alpha)}\right)} \tag{7}$$

Substituting the solution for $l$ into (5), we get

$$ \frac{Xw}{p} = \frac{(1 - \alpha)}{\alpha} \, c \, \left(1 + \frac{\alpha}{(1 -\alpha)}\right)$$

<br/>

$$ \frac{Xw}{p} = \frac{(1 - \alpha)}{\alpha} \, \left(c + \frac{c \, \alpha}{1 - \alpha}\right) $$

<br/>

$$ \frac{Xw}{p} = \frac{(1 - \alpha)}{\alpha} \,  c + c$$

<br/>

$$ c = \frac{Xw}{p} \, \frac{1}{\frac{1 - \alpha}{\alpha} + 1} \tag{8}$$

So, the optimal $(c,l) = ((8), (7))$

### 2)

Consider the utility function $$u() = 2\sqrt{c} + 2\sqrt{l}$$

<br/>

##### Repeat (b) - (e) with this utility function.

#### b)

$$L() = 2\sqrt{c} + 2\sqrt{l} + \lambda(Xw - Xwl - pc)\tag{10}$$

<br/>

#### c)

$$ \frac{\partial L()}{\partial c} = \frac{1}{\sqrt{c}} - \lambda p$$

<br/>

$$ \frac{\partial L()}{\partial l} = \frac{1}{\sqrt{l}} - \lambda Xw$$

<br/>

$$ \frac{\partial L()}{\partial \lambda} = Xw - Xwl - pc \tag{11}$$

<br/>

#### d)

$$ \left( \frac{l}{c} \right)^{0.5} = \frac{p}{Xw}$$

<br/>

$$ c = \left( \frac{Xw}{p} \right) ^2 l \tag{12}$$

Plugging this into our budget constraint, we get
<br/>

$$ Wx - Wxl - \frac{Xw^2l}{p} = 0 $$ 

<br/>

$$ Wx( 1 - l - \frac{Xwl}{p}) = 0 $$

<br/>

$$ 1 = l(1 + \frac{Xw}{p}) $$

<br/>

$$ l = \frac{1}{1 + \frac{Xw}{p}} \tag{13}$$

<br/>

Plugging this back into $(12)$, we get

<br/>

$$ c = \frac{X^2w^2}{p}\frac{1}{p + Xw} $$

<br/>

$$ \bar{c} = \frac{X^2w^2}{p^2 + pXw} $$

which transforms into

$$ \bar{c} = \frac{\left(\frac{Xw}{p}\right)^2}{1 + \frac{Xw}{p}} $$

So,


$$ \bar{l} = \frac{1}{1 +\frac{Xw}{p}} , \,\bar{c} = \frac{\left(\frac{Xw}{p}\right) ^2}{1 + \frac{Xw}{p}}$$

### 3)

Now suppose the consumer is required to pay a lump-sum tax $T > 0$, independented of the consumer's income. Assume the utility function $$ u(c,l) = 2\sqrt{c} + 2\sqrt{l} $$and assume $w > T$, so that the consumer can afford to pay the tax if she supplies enough labor. Study the effect of the lump-sum tax on the labor supply curve, using the following steps:



#### a)
##### Graph the consumer's budget constraint when there is no tax and where there is a positive tax

The consumer's single budget constraint is 

$$ w(1-l) - T = pc \tag{13}$$


Graphically:

$$c = \frac{w}{p} - \frac{wl}{p} - \frac{T}{p} $$

```python
cval3  = []
cval3t = []
T = rand.randint(0,10)
w = rand.randint(10, 20)
p = rand.randint(1,20)
rwage = w/p
```

```python
for l in lval:
    c = rwage- rwage*l
    cval3.append(c)
    c = rwage - rwage*l - T/p
    cval3t.append(c)
    
plt.plot(lval, cval3, label = 'Tax = 0')
plt.plot(lval, cval3t,label = 'Tax = {}'.format(T))
plt.ylabel('c')
plt.xlabel('l')
plt.title('Budget constraint with w/p = {}'.format(round(rwage, 2)))
plt.legend(loc='upper right')
plt.show()
```

#### b)

###### Write down the consumer's optimization problem

$$ L() = \alpha \, \ln(c) + (1 - \alpha) \, \ln(l) + \lambda(w - wl - T - pc)  \tag{14}$$

#### c)

##### Derive the first-order conditions


<br/>


$$ \frac{\partial L()}{\partial c} = \frac{\alpha}{c} - \lambda p\tag{15}$$

<br/>

$$ \frac{\partial L()}{\partial l} = \frac{(1 - \alpha)}{l} - \lambda w\tag{16}$$

<br/>

$$ \frac{\partial L()}{\partial \lambda} = w - wl - T - pc\tag{17}$$

#### d)

##### Solve to find the optimal $(c,l)$


Eliminating the $\lambda$, we get 

$$ \frac{w}{p} = \frac{c \, (1 - \alpha)}{l \, \alpha} \tag{18}$$

Solving for $cp$, we get 
$$ cp = wl \frac{\alpha}{(1 - \alpha)} \tag{19} $$

<br/>

Plugging this into our constraint, we get

$$w - wl - T - wl \frac{\alpha}{(1- \alpha)} = 0 $$

<br/>

$$ w - T = wl +  wl \frac{\alpha}{(1 - \alpha)} $$

<br/>

$$ wl \, (1 + \frac{\alpha}{1 - \alpha}) = w - T $$

<br/> 

$$ wl = \frac{w - T}{\frac{1}{(1 - \alpha)}}$$

$$ \bar{wl} = (w - T)(1 - \alpha)$$

<br/>

$$ \bar{l} = (1 - \frac{T}{w})(1 - \alpha) \tag{20}$$

<br/>

Plugging this back into $(18)$, we find $\bar{c}$

$$\bar{cp} = \frac{\alpha}{(1 - \alpha)} (w - T)(1 - \alpha) = \alpha(w - T) $$

<br/>

$$ \bar{c} = \frac{\alpha(w - T)}{p} \tag{21}$$ 

So, our optimal (c,  l) is 

$$ \bar{c} = \frac{\alpha(w - T)}{p}, \, \bar{l} = (1 - \frac{T}{w})(1 - \alpha)$$ 

#### e)

#### Graph the consumer's labor supply function

Given $n+l = 1$, we find that

$$ n = \frac{\alpha w + (1 - \alpha) T}{w}$$

To express this in terms of real wage, we divide the numerator and denominator by $p$

$$ n = \frac{\alpha \frac{w}{p} + (1 - \alpha) \frac{T}{p}}{\frac{w}{p}} $$

$$ = \alpha + \frac{1 - \alpha \frac{T}{p}}{\frac{w}{p}} $$

```python
T = rand.randint(0,10)
p = rand.randint(1,20)
nval = np.linspace(0.601, 1, 500)
```

```python
rwagelst = []
for n in nval:
    rwage = (0.6)/(n-0.4)* (T/p)
    rwagelst.append(rwage)
    
plt.plot(nval, rwagelst)
plt.ylabel('w/p')
plt.xlabel('n')
plt.title('Labor supply when alpha = 0.6')
plt.show()
```

We can see that when the real wage is $T/p$, the consumer has to spend all her time working in order to pay the tax, as $n = 1$. However, her labor supply, that is how much time she spends workings, will decrease as her real wage increases. As real wage increases, the lump-sum tax becomes smaller, and her labor supply converges to $\alpha$.

Even though the consumer has the same utility function as in question 1, the lump-sum tax creates a backward-bending labor supply curve.
